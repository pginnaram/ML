**Objectives:**  Here we visually explore how the GD actually descends using a one Dimensional Quadratic Error Function. We explore the rate and manner of descent with changing learning rate.

**Expected Time:** This Experiment should take around 20 mins

```python
import matplotlib
matplotlib.use('nbagg')
import numpy as np
import matplotlib.pyplot as plt
```

eta = 0.2

```python
def error(m): return (m**2) + (2 * m) + 2
def gradient(m): return 2*m + 2
def delta(m, eta): return eta*gradient(m)

#Set the initial parameter
m = 5
# Set the learning rate
eta = 0.2

nb_of_iterations = 20  # number of gradient descent updates
w_err = [np.array([m, error(m)])] # List to store the m, error values
for i in range(nb_of_iterations):
    dm = delta(m, eta)  # Get the delta m update
    m = m - dm  # Update the current m value
    w_err.append(np.array([m, error(m)]))  # Add m,error to list

# Print the final w, and cost
#for i in range(0, len(w_err)):
    #print('w({}): {:.4f} \t cost: {:.4f}'.format(i, w_err[i][0], w_err[i][1]))
w_err=np.array(w_err)
#print(w_err[:,0])
#plt.plot(w_err[:,0],w_err[:,1])
#plt.show()
```

```python
def error(m): return (m**2) + (2 * m) + 2
def gradient(m): return 2*m + 2
def delta(m, eta): return eta*gradient(m)

#Set the initial parameter
m = 5
# Set the learning rate
eta = 0.5

nb_of_iterations = 20  # number of gradient descent updates
w_err2 = [np.array([m, error(m)])] # List to store the m, error values
for i in range(nb_of_iterations):
    dm = delta(m, eta)  # Get the delta m update
    m = m - dm  # Update the current m value
    w_err2.append(np.array([m, error(m)]))  # Add m,error to list

# Print the final w, and cost
for i in range(0, len(w_err2)):
    print('w({}): {:.4f} \t cost: {:.4f}'.format(i, w_err2[i][0], w_err2[i][1]))
w_err2=np.array(w_err2)
```

```python
def error(m): return (m**2) + (2 * m) + 2
def gradient(m): return 2*m + 2
def delta(m, eta): return eta*gradient(m)

#Set the initial parameter
m = 5
# Set the learning rate
eta = 0.7

nb_of_iterations = 20  # number of gradient descent updates
w_err1 = [np.array([m, error(m)])] # List to store the m, error values
for i in range(nb_of_iterations):
    dm = delta(m, eta)  # Get the delta m update
    m = m - dm  # Update the current m value
    w_err1.append(np.array([m, error(m)]))  # Add m,error to list

# Print the final w, and cost
for i in range(0, len(w_err1)):
    print('w({}): {:.4f} \t cost: {:.4f}'.format(i, w_err1[i][0], w_err1[i][1]))
w_err1=np.array(w_err1)
```

```python
x = range(-10,10)
m = []
for i in x:
    m.append(error(i))
```

```python
plt.grid(True)
plt.plot(x,m)
plt.plot(w_err[:,0],w_err[:,1],"o")
plt.plot(w_err1[:,0],w_err1[:,1],"o")
plt.plot(w_err2[:,0],w_err2[:,1],"o")
n = range(1,len(w_err[:,0]))
plt.legend(["x vs m","eta = 0.2","eta = 0.7","eta = 0.5"])
for i, txt in enumerate(n):
    plt.annotate(txt, (w_err[:,0][i],w_err[:,1][i]))
    plt.annotate(txt, (w_err1[:,0][i],w_err1[:,1][i]))
    plt.annotate(txt, (w_err2[:,0][i],w_err2[:,1][i]))
plt.show()
```
