### Applying non-linear Regression


#### Importing the required libaries

```python
import numpy as np
from scipy.optimize import curve_fit
## Assuming some range of  "X" values
X = np.arange(-10,10, 0.5)
```

```python
#Fucntion to calculate y values
def value(x):
    return 3 * x**2 + 2 * x + 1 
```

#### Adding some noise

```python
import random
## adding some noise to the y values using random function
y = value(X) * (random.random() * 0.2 + 0.9)
```

```python
#plotting the graph
import matplotlib.pyplot as plt
plt.plot(X,y, ".")
plt.show()
```

#### Try to fit the line on the data 

```python
import scipy.stats as stat
# applying linear regression
m, c, _, _, _ = stat.linregress(X, y)
#Calculating the predicted values
ylr = m * X + c 
plt.figure(figsize=(16,10))
# plotting the data
plt.plot(X, y, 'o')
# plotting the line
plt.plot(X, ylr)
plt.show() 
```

#### We can see that the line is not properly fitting into data

#### Now We will use some mappimg function then we try to fit the line on that data.
#### Mapping  = (x , y ) to (x ** 2 , y)

```python
m, c, _, _, _ = stat.linregress(X * X, y)
ylr = m * X * X + c 
plt.figure(figsize=(16,10))
plt.plot(X * X, y, 'o')
plt.plot(X * X, ylr)
plt.show() 
```
